```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
df=pd.read_csv("/content/Salary_Data.csv")
df
```

```python
x=df.iloc[:,:-1]
x
```

```python
y=df.iloc[:,-1]
y
```

```python
plt.scatter(x,y,color="r")
```

```python
plt.xlabel('Year of Experience')
plt.ylabel('Salary')
plt.title('Experience-Salary Graph',size=15,color="g")
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
df1=pd.DataFrame({'Actual_value':y_test,'Predicted_value':y_pred,'Difference':y_test-y_pred})
df1
```

```python
print(model.predict([[5.8]]))
```

```python
plt.scatter(x_train,y_train)
plt.plot(x_test,y_pred,color="r")
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.title("Experience-Salary Graph",fontsize=15,fontweight="bold",fontstyle="italic")
```

```python
y_test
```

```python
print("Slop is",model.coef_)
```

```python
print("constant is",model.intercept_)
```

```python
from sklearn.metrics import mean_absolute_error
print("MAE is",mean_absolute_error(y_test,y_pred))
```

```python
from sklearn.metrics import mean_absolute_percentage_error
print("MAPE is",mean_absolute_percentage_error(y_test,y_pred))
```

```python
from sklearn.metrics import mean_squared_error
print("MSE is",mean_squared_error(y_test,y_pred))
```

```python
from sklearn.metrics import root_mean_squared_error
print("RMSE is",root_mean_squared_error(y_test,y_pred))
```

```python
from sklearn.metrics import r2_score
print("score is",r2_score(y_test,y_pred))
```
